```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)
arr2 = arr  # alias, shallow copy 
arr2[0] = [1, 2, 3, 4, 5]
print('arr2 \n', arr2)
print('arr \n', arr)  # also CHANGED !! -- same as deep copy
x = 5 # 5
y = x 
y = y + 1 # 6
# next cell
```

```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)
arr2 = arr.copy()  # deep copy
arr2[0] = [1, 2, 3, 4, 5]
print('arr2 \n', arr2)
print('arr \n', arr)  # not CHANGED !! -- same as deep copy
```

```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(5))
print(arr)
print('arr > 55 \n', arr > 55)  # condition -> creates bool array
print()
print('arr[arr > 55]', arr[arr > 55])
```

```python
# 1 create 20 numbers between 0-100 (seed 42) and keep only even numbers
# 2 create 20 numbers between 0-100 (seed 42) and keep only number smllaer than 40
# 3 arr = np.array([10, 20, 30, 99, 40, 50, 99, 60]) keep only numbers not 99
```

```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(10))
print(arr)
print('arr > 55 \n', (arr >= 51) & (arr <= 71) )  # condition -> between 51 and 71
print()
print('arr[71 >= arr >= 55]', arr[(arr >= 51) & (arr <= 71) ])
print('arr[71 < arr or arr < 55]', arr[(arr < 55) | (arr > 71) ])
```

```python
np.random.seed(42)
arr = np.random.randint(0, 100, size=(5, 5))
print(arr)
print('max in matrix', arr.max())
print('max per column', arr.max(axis=0))
print('max per row', arr.max(axis=1))
print('min', arr.min())
print('arg_max index-max per column #', arr.argmax(axis=0))
print('arg_min index-min per column #', arr.argmin(axis=0))
print('sum per column', arr.sum(axis=0))
print('sum per row', arr.sum(axis=1))
print(arr.shape)
```
